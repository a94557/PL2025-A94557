```python
import re

def extrair_compositores(ficheiro):
    compositores = []

    with open(ficheiro, 'r', encoding='utf-8') as f:
        conteudo = f.readlines()


    for linha in conteudo[1:]:  
        campos = re.split(r';', linha.strip())
        if len(campos) == 7:  
            compositor = campos[4].strip()  
            compositores.append(compositor)

    
    compositores = sorted(set(compositores)) 

    for compositor in compositores:
        print(compositor)

if __name__ == "__main__":
    extrair_compositores('obras.csv')
```

```python
import re

def contar_obras_por_periodo(ficheiro):
    contagem_periodos = {}

    with open(ficheiro, 'r', encoding='utf-8') as f:
        conteudo = f.readlines()

    for linha in conteudo[1:]: 
        campos = re.split(r';', linha.strip())
        if len(campos) == 7: 
            periodo = campos[3].strip() 
            if periodo: 
                if periodo in contagem_periodos:
                    contagem_periodos[periodo] += 1
                else:
                    contagem_periodos[periodo] = 1

    print("Distribuição das obras por período:")
    for periodo, contagem in contagem_periodos.items():
        print(f"{periodo}: {contagem} obras")

if __name__ == "__main__":
    contar_obras_por_periodo('obras.csv')
```

```python
import re

def dicionario_obras_por_periodo(ficheiro):
    periodo_obras = {}

    with open(ficheiro, 'r', encoding='utf-8') as f:
        conteudo = f.readlines()

    for linha in conteudo[1:]:  
        campos = re.split(r';', linha.strip())
        if len(campos) == 7: 
            periodo = campos[2].strip()  
            titulo = campos[0].strip()  
            
            if periodo: 
                if periodo not in periodo_obras:
                    periodo_obras[periodo] = []
                periodo_obras[periodo].append(titulo)

    for periodo in periodo_obras:
        periodo_obras[periodo] = sorted(set(periodo_obras[periodo]))  

    print("Dicionário de obras por período:")
    for periodo, titulos in periodo_obras.items():
        print(f"{periodo}: {titulos}")

if __name__ == "__main__":
    dicionario_obras_por_periodo('obras.csv')
```
